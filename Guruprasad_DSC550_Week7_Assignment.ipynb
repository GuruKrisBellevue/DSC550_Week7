#### 1.7. Run a linear regression and report the R2-value and RMSE on the test set.

```python
# Creating a linear regression object
linear_regress = LinearRegression()
```

```python
# training the linear regression model on the training data
linear_regress.fit(X_train, Y_train)
# Calculating the score of the model
print(linear_regress.score(X_test, Y_test))
```

```python
# Creating a function to calculate the metrics
def create_regression_metrics(y_obj,predicted_tgt):
    """
    This function takes the Predicted values as input and returns the results of various metrics like R2, 
    MAE, MSE and RMSE
    """
    # using r2_score to calculate the R2
    r2=r2_score(y_obj, predicted_tgt)
    print(f"R2: {r2} ")
    # using mean_absolute_error functionto calculate the MAE
    mae=mean_absolute_error(y_obj,predicted_tgt)
    print(f"MAE: {mae}")
    # using mean_squared_error function to calculate the MSE
    mse=mean_squared_error(y_obj,predicted_tgt)
    print(f"MSE: {mse}")
    # using mean_squared_error function to calculate the RMSE by taking square root of MSE
    rmse= np.sqrt(mean_squared_error(y_obj,predicted_tgt))
    print(f"RMSE: {rmse}")
    return (r2,mae,mse,rmse)
```

```python
# Creating a variable for Predicted value for Linear Regression Test set
pred_test_olr=linear_regress.predict(X_test)
# Getting the Metrics for Ordinary Linear Regression Test set
r2_test_olr,mae_test_olr,mse_test_olr,rmse_test_olr=create_regression_metrics(Y_test,pred_test_olr)
```
